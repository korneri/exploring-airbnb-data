Show the listings per neighbourghood, with the most popular neighbourhoods first (that is, in decreasing neighbourhood popularity order).

## Availability per Day

Create a plot with the number of listings available per day, on the $y$ axis, and the date, on the $x$ axis.

```python
#reading calendar.csv file
calendar=pd.read_csv("D:\\Users\\putyourpath\\airbnb\\calendar.csv\\calendar.csv")

#grouping by date and countings only the listing that have availability on that date (available == 't')
listPerDate = pd.DataFrame(calendar[calendar['available'] == 't'].groupby('date').size())
#renaming the column as ListingsPerDate
listPerDate = listPerDate.rename(columns = {listPerDate.columns[0]:'ListingsPerDate'})
#getting the sum of listings per date
listingsPerDate = listPerDate[['ListingsPerDate']]
#getting the dates as a list
dates= list(listPerDate.index)
#converting dates to datetime type 
dates = pd.to_datetime(dates)

#creating the plot with thr number of listings available per date
plt.figure(figsize=(14, 7))
plt.plot(dates, listingsPerDate)
#naming the x and y axis 
plt.title('Number of Available Listings Per Date')
plt.ylabel('Number Of Available Listings')
plt.xlabel('Dates')
#showing the plot
plt.show()
```

## Reviews, Occupacy per Listing, Average Income

Create a histogram showing the number of reviews per listing. The histogram should bin the number of reviews, from the smallest amount of reviews in a listing, to the maximum number of reviews in a listing rounded up to the closest hundred.


```python
#reading reviews.csv file
reviews=pd.read_csv("D:\\Users\\putyourpath\\airbnb\\reviews.csv~1\\reviews.csv")
#grouping by listing_id, count of reviews per listing     
reviewsPerListing = pd.DataFrame(reviews['id'].groupby(reviews['listing_id']).size()) 
#renaming the column as ReviewsPerListing
reviewsPerListing = reviewsPerListing.rename(columns = {reviewsPerListing.columns[0]:'ReviewsPerListing'})
#rounding in the closest hundred and converting into int
reviewsPerListing['ReviewsPerListing']=  (round(reviewsPerListing[['ReviewsPerListing']] / 1000 ,1) * 1000).astype(int)

#sorting by  number of listings in descending order
reviewsPerListing = reviewsPerListing.sort_values(by=['ReviewsPerListing'],ascending=False)
 
#creating the histogram
reviewsPerListing.hist(color='turquoise',edgecolor='black', 
                  figsize=(17, 5))
#naming the x and y axis
plt.xlabel('Listing_Ids')
plt.ylabel('Number of Reviews')
plt.title('Number of Reviews Per Listing_id')
#showing the histogram
plt.show()

```

Once you have done that, calculate the average occupacy per listing per month. To calculate that, find the average reviews per month and assume that half of the bookings result in reviews.

```python
#getting the listing_id and id column
reviewsPerListingPerYPerM = reviews[['listing_id','id']]
#getting year and month from date using datetime
reviewsPerListingPerYPerM['yearOfReviewsPerListing'] = pd.to_datetime(reviews['date']).dt.year
reviewsPerListingPerYPerM['monthOfReviewsPerListing'] = pd.to_datetime(reviews['date']).dt.month
#grouping by listing_id, year, month, finding the average reviews 
reviewsPerListingPerYPerM = reviewsPerListingPerYPerM.groupby(['listing_id','yearOfReviewsPerListing','monthOfReviewsPerListing']).mean()
#renaming the column to AverageReviewsPerListing
reviewsPerListingPerYPerM = reviewsPerListingPerYPerM.rename(columns = {reviewsPerListingPerYPerM.columns[0]:'AverageReviewsPerListing'})
# average occupacy per listing per month, multipling to 2 assuming that half of bookings result in reviews
reviewsPerListingPerYPerM['AverageOccupancyPerListingPerMonth'] = reviewsPerListingPerYPerM['AverageReviewsPerListing']*2
reviewsPerListingPerYPerM
```

Then, assuming that each booking is for a three nights stay, calculate the average income per month, using the average price charged throughout all listings.

```python

#putting out dollar and , symbols , converting the number into a float type
calendar['price'] = calendar['price'].str.strip('$').str.replace(',', '')
calendar['price'] = calendar['price'].astype(float)

#getting only the column interested in this subquery,multiplying price with  3 assuming that
#each booking is for a three nights stay
avgPricePerMonth = calendar[['price']]*3
#getting year and month from date
avgPricePerMonth['yearOfBookings'] = pd.to_datetime(calendar['date']).dt.year
avgPricePerMonth['monthOfBookings'] = pd.to_datetime(calendar['date']).dt.month
#grouping price by year and month , taking the mean of price
avgPricePerMonth = avgPricePerMonth.groupby(['yearOfBookings','monthOfBookings']).mean()
#renaming the column to AveragePricePerBooking
avgPricePerMonth = avgPricePerMonth.rename(columns = {avgPricePerMonth.columns[0]:'AveragePricePerBooking'})
#getting column listing_id
listingsPerMonth = calendar[['listing_id']]
#getting the year and month of date
listingsPerMonth['yearOfListings'] = pd.to_datetime(calendar['date']).dt.year
listingsPerMonth['monthOfListings'] = pd.to_datetime(calendar['date']).dt.month
#rouping listing_id by year and month, calculating the count of listings 
listingsPerMonth = pd.DataFrame(listingsPerMonth.groupby(['yearOfListings','monthOfListings']).size())
#renaming the column as NumberOfListingsPerMonth
listingsPerMonth = listingsPerMonth.rename(columns = {listingsPerMonth.columns[0]:'NumberOfListingsPerMonth'})
#calculating the average price per month which is average price per booking multiblied with number of listings per month
avgPricePerMonth['AverageIncomePerMonth'] = avgPricePerMonth['AveragePricePerBooking']*listingsPerMonth['NumberOfListingsPerMonth']

avgPricePerMonth
```

## Listings per Host

Create a table with the number of listings per host, in descending order. Create a graph that will show the distribution of listings per host, as follows:

* You will rank the host on a number of listings basis. So, the first host will be the one with the largest number of listings, the second host will be the one with the second largest number of listings, etc. If there is a number of hosts with the same listings, just add them in sequence. For instance, if there are 10 hosts with one listing each, add them in the ranking with any order between them.

* On the $y$ axis you will have the number of listings.

That is, a point $(x, y)$ will mean that the $x$-th highest hosting host has $y$ listings.

```python
#grouping listing id by host id, calculating the number of listings per host
listingsPerHost = pd.DataFrame(listings['id'].groupby(listings['host_id']).size())
#renaming the column as NumberOfListingsPerHost
listingsPerHost = listingsPerHost.rename(columns = {listingsPerHost.columns[0]:'NumberOfListingsPerHost'})
#putting number of listings per host in descending order
listingsPerHost = listingsPerHost.sort_values(by=['NumberOfListingsPerHost'],ascending=False)
#creating a ranking column based on number of listings per host in descending row 
listingsPerHost['Ranking'] = listingsPerHost['NumberOfListingsPerHost'].rank(method='first', ascending=0)
#psorting elements in ascending order based on ranking
listingsPerHost = listingsPerHost.sort_values(by=['Ranking'],ascending=True)

#getting NumberOfListingsPerHost
listing_num = listingsPerHost[['NumberOfListingsPerHost']]
#getting Ranking
ranking= listingsPerHost[['Ranking']]

#creating the plot containing the variables got above
plt.figure(figsize=(15, 7))
plt.plot(ranking, listing_num)
#setting the names of x and y axis
plt.title('Number of listings per host_id')
plt.ylabel('Number of Listings')
plt.xlabel('Ranking')
#show the plot
plt.show()
```

## Visualize Listings

Draw an interactive map to show the listings ovelayed on Athens.

You can use the [folium](https://github.com/python-visualization/folium) library to create your map; spend some time reading the documentation.

The map should show each listing and on clicking on top of it, the user should see a short description of the listing.

Note that such a map is heavy on resources. You will probably *not* be able to include it in a Jupyter notebook, so you should save it to an external HTML file that can be opened by any browser. Also, to make things easier, you make wish to limit the maximum zoom level from 12 to 15 (after you read folium's documentation you will know what this is).

Hint: be careful witht he description data, if they contain some characters they may not be rendered in HTML and may destroy your output.

```python
#getting id, latitude, lingitude, neighbourhood, price columns
points = listings[['id','latitude', 'longitude','neighbourhood','price']]

#setting the map size based on the average of longitude and latitude,setting zoom_start = 12 and max_zoom = 15
map = folium.Map(location=[points['latitude'].mean(), 
                                points['longitude'].mean()], 
                      zoom_start=12,max_zoom = 15)

#creating a Circle Marker for each point, each point will get a popup 
#popup will show the listing is, neighbourhood and price of listing
for row in points.itertuples():
     map.add_child(folium.CircleMarker(location=[row.latitude,row.longitude], radius=0.5,
                 color='#0080bb', fill_color='#0080bb',
             popup=('Listing_id: ' + str(row.id) + '<br>' 
                   'Neighbourhood: ' + str(row.neighbourhood) + '<br>'
                   
                  'Price: ' + str(row.price) + ' $'
                    )))
#saving the map as an html file
map.save("map.html")
```
